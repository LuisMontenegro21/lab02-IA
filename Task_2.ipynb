```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import re
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
def clean(mjs):
    mjs = re.sub(r'[^a-zA-Z0-9\s]', '', mjs.lower())
    return mjs
```

```python
# Create clasification diccionary 
data = {
    "spam" : {"emails": 0, "words": {}}, 
    "ham": {"emails": 0, "words": {}}
}
```

```python
def readlines(line, data):
    tag, mjs = line.strip().split('\t', 1)
    newmjs = clean(mjs)
    if tag not in data:
        data[tag] = {"emails": 0, "words": {}}
    data[tag]["emails"] += 1
    for word in newmjs.split():
        data[tag]["words"][word] = data[tag]["words"].get(word, 0) + 1 
```

```python
with open('entrenamiento.txt', 'r') as file:
        for line in file:
            readlines(line, data)
```

## Task 2.2

```python
import numpy as np

def bayes(data, alpha=1):
    spam = data["spam"]["emails"]
    ham = data["ham"]["emails"]
    spam_words = sum(data["spam"]["words"].values())
    ham_words = sum(data["ham"]["words"].values())

    p_spam = spam / (spam + ham)
    p_ham = ham / (spam + ham)

    probability = {
        "p_spam": p_spam,
        "p_ham": p_ham,
        "p_spam_prob": {},
        "p_ham_prob": {}
    }

    for word in set(data["spam"]["words"].keys()).union(set(data["ham"]["words"].keys())):
        word_occurrence_spam = data["spam"]["words"].get(word, 0) + alpha
        word_occurrence_ham = data["ham"]["words"].get(word, 0) + alpha

        p_word_spam = word_occurrence_spam / (spam_words + len(set(data["spam"]["words"].keys())) * alpha)
        p_word_ham = word_occurrence_ham / (ham_words + len(set(data["ham"]["words"].keys())) * alpha)

        probability["p_spam_prob"][word] = p_word_spam
        probability["p_ham_prob"][word] = p_word_ham

    return probability

# Ejemplo de uso
ejemp = bayes(data)
print("Modelo entrenado:", ejemp)
```

se genera un imput para que el usario escriba el mensaje analizar y muestra al final si es ham o spam
mensajes como: "can you come to my house please", "call us for free wifi"

```python
def classify_new_message(message, model):
    cleaned_message = clean(message)
    words = cleaned_message.split()

    p_spam = model["p_spam"]
    p_ham = model["p_ham"]

    for word in words:
        p_spam *= model["p_spam_prob"].get(word, 1)
        p_ham *= model["p_ham_prob"].get(word, 1)

    total_probability = p_spam + p_ham

    probability_spam = p_spam / total_probability
    probability_ham = p_ham / total_probability

    decision = "spam" if probability_spam > probability_ham else "ham"

    return probability_spam, probability_ham, decision

# Solicitar al usuario que ingrese un mensaje
nuevo_mensaje = input("Ingrese el mensaje que desea verificar: ")
prob_spam, prob_ham, decision = classify_new_message(nuevo_mensaje, ejemp)

print("Probabilidad Spam:", prob_spam)
print("Probabilidad Ham:", prob_ham)
print("Decisin:", decision)
```
